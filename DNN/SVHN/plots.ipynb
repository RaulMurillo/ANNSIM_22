```python
import numpy as np
import matplotlib.pyplot as plt
import json
```

```python
# FP32
dir = './FP32/'
with open(dir+'train.hist', 'r') as f_in:
    train_hist = json.load(f_in)
with open(dir+'test.hist', 'r') as f_in:
    test_hist = json.load(f_in)
fp32_train = train_hist
fp32_test  = test_hist

# FP16
dir = './FP16/'
with open(dir+'train.hist', 'r') as f_in:
    train_hist = json.load(f_in)
with open(dir+'test.hist', 'r') as f_in:
    test_hist = json.load(f_in)
fp16_train = train_hist
fp16_test  = test_hist

# bfloat16
dir = './bfloat16/'
with open(dir+'train.hist', 'r') as f_in:
    train_hist = json.load(f_in)
with open(dir+'test.hist', 'r') as f_in:
    test_hist = json.load(f_in)
bfloat16_train = train_hist
bfloat16_test  = test_hist

# P8
dir = './P8/'
with open(dir+'train.hist', 'r') as f_in:
    train_hist = json.load(f_in)
with open(dir+'test.hist', 'r') as f_in:
    test_hist = json.load(f_in)
P8_train = train_hist
P8_test  = test_hist

# P10
dir = './P10/'
with open(dir+'train.hist', 'r') as f_in:
    train_hist = json.load(f_in)
with open(dir+'test.hist', 'r') as f_in:
    test_hist = json.load(f_in)
P10_train = train_hist
P10_test  = test_hist

# P12
dir = './P12/'
with open(dir+'train.hist', 'r') as f_in:
    train_hist = json.load(f_in)
with open(dir+'test.hist', 'r') as f_in:
    test_hist = json.load(f_in)
P12_train = train_hist
P12_test  = test_hist

# P14
dir = './P14/'
with open(dir+'train.hist', 'r') as f_in:
    train_hist = json.load(f_in)
with open(dir+'test.hist', 'r') as f_in:
    test_hist = json.load(f_in)
P14_train = train_hist
P14_test  = test_hist

# P16
dir = './P16/'
with open(dir+'train.hist', 'r') as f_in:
    train_hist = json.load(f_in)
with open(dir+'test.hist', 'r') as f_in:
    test_hist = json.load(f_in)
P16_train = train_hist
P16_test  = test_hist
```

```python
# Plotted the accuracy Graph
fig= plt.figure(figsize=(7,5))

all_test_hist = [fp32_test, fp16_test, bfloat16_test, P16_test, P14_test, P12_test, P10_test, P8_test]
names = ['Float32', 'Float16', 'BFloat16', 'Posit16', 'Posit14', 'Posit12', 'Posit10', 'Posit8']

for i, test in enumerate(all_test_hist):
    accuracies = [x['accuracy'] for x in test]
    plt.plot(accuracies, label=names[i])

plt.legend()
plt.xlabel('epoch')
plt.ylabel('accuracy %')
# plt.title('Accuracy vs. No. of epochs')

plt.savefig('accuracy_svhn.pdf', bbox_inches='tight')
plt.show()
```

```python
# Plotted the loss Graph
fig= plt.figure(figsize=(7,5))

all_test_hist = [fp32_test, fp16_test, bfloat16_test, P16_test, P14_test, P12_test, P10_test, P8_test]
names = ['Float32', 'Float16', 'BFloat16', 'Posit16', 'Posit14', 'Posit12', 'Posit10', 'Posit8']

for i, test in enumerate(all_test_hist):
    losses = [x['loss'] for x in test]
    plt.plot(losses, label=names[i])

plt.legend()
plt.xlabel('epoch')
plt.ylabel('loss')
# plt.title('Loss vs. No. of epochs')

plt.savefig('loss_svhn.pdf', bbox_inches='tight')
plt.show()
```
